```python
df = df[['id', 'date_joined', 'username']]
df = df.rename(columns={'id':'user_id', 'date_joined':'created_at', 'username': 'login'})
```

```python
df = df.dropna().replace(r'\$', '', regex=True)
df
```

```python
dates = list(pd.to_datetime(df['created_at']).dt.strftime('%Y-%m-%d').unique())
dates
```

```python
day_week = pd.to_datetime(df['created_at']).dt.day_name().value_counts().reset_index()
registrations_by_weekday = dict(zip(day_week.created_at, day_week['count']))
registrations_by_weekday
```
