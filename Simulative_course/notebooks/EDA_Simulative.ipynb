```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('D:/GitHub_projects/Simulative_course/data/data.csv', encoding='1251')
```

```python
pd.set_option('display.max_columns', 40)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
our_cols = ['DR_Dat', 'DR_Tim', 'DR_NChk', 'DR_NDoc', 'DR_Apt', 'DR_NDrugs', 'DR_Kol','DR_CZak', 'DR_CRoz', 'DR_SDisc',
             'DR_TPay', 'DR_CDrugs', 'DR_Suppl', 'DR_CDisc', 'DR_BCDisc', 'DR_TabEmpl','DR_VZak', 'DR_Pos']
```

```python
df = df[our_cols]
```

```python
df.columns = ['dt', 'c_time','nchk', 'ndoc', 'apt', 'drug', 'kol','zak', 'roz', 'disc',
             'pay_type', 'drug_id', 'suppl', 'disc_id', 'disc_barcode', 'empl', 'vzak', 'pos']
```

```python
df.loc[:, 'disc_barcode'] = df['disc_barcode'].astype(str).str.replace(r'\.0$', '', regex=True)
```

```python
df.loc[:, 'disc_id'] = df['disc_id'].astype(str).str.replace(r'\.0$','', regex=True)
```

```python
df.loc[:, 'dt'] = pd.to_datetime(df['dt']).dt.strftime('%d.%m.%Y')
```

```python
df.loc[:, 'vzak'] = df['vzak'].astype(str).replace({'1': 'Обычный', '2': 'Интернет'})
```
