```python
df.columns = ['dt', 'c_time','nchk', 'ndoc', 'apt', 'drug', 'kol','zak', 'roz', 'disc',
             'pay_type', 'drug_id', 'suppl', 'disc_id', 'disc_barcode', 'empl', 'vzak', 'pos']
```

```python
df.dtypes
```

```python
#Изменение типа данных столбца
df.loc[:, 'disc_barcode'] = df['disc_barcode'].astype(str).str.replace(r'\.0$', '', regex=True)
```

```python
df.loc[:, 'disc_id'] = df['disc_id'].astype(str).str.replace(r'\.0$','', regex=True)
```

```python
#Изменение дат
df.loc[:, 'dt'] = pd.to_datetime(df['dt']).dt.strftime('%d.%m.%Y')
```

```python
#Изменить тип данных у вида закупки
#df['vzak'].unique()   ## Проверка уникальных значений
df.loc[:, 'vzak'] = df['vzak'].astype(str).replace({'1': 'Обычный', '2': 'Интернет'})
```
