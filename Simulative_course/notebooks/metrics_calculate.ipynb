```python
grouped_df = df.groupby('DR_NDrugs').agg({'DR_Kol' : 'sum'})
grouped_df
```

Многомерный ABC анализ

```python
grouped2_df = df.groupby('DR_NDrugs').agg({'DR_Kol' : 'sum', 'DR_CRoz' : 'sum'})
grouped2_df
```

```python
grouped2_df['real_kol'] = (grouped2_df['DR_Kol']/sum(grouped2_df['DR_Kol'])).sort_values(ascending=False)
grouped2_df
```

Функция ABC анализа по датафрейму и нужным столбцам

```python
df1 = df[['DR_NDrugs', 'DR_Kol', 'DR_CRoz']]
df1.columns
```

```python
#Пример удаления значения
a = list(df1.columns)
a.remove('DR_Kol')
a
```

```python
def perform_abs(df, index):
    """
    На вход подается датафрейм Pandas...
    """
    cols = list(df.columns)
    cols.remove(index)
    grouped_df = df.groupby(index).agg({col : 'sum' for col in cols})
    for col in cols:
        grouped_df[f'real_{col}'] = grouped_df[col] / sum(grouped_df[col])
        grouped_df = grouped_df.sort_values(f'real_{col}', ascending=False)
        grouped_df[f'cumsum_{col}'] = grouped_df[f'real_{col}'].cumsum()
        grouped_df[f'abc_{col}'] = np.where(grouped_df[f'cumsum_{col}'] < 0.8, 'A', np.where(grouped_df[f'cumsum_{col}'] < 0.95, 'B', 'C'))
    return grouped_df[cols + [f'abc_{col}' for col in cols]]
```

```python
df_xyz = df_xyz.groupby(['DR_Dat', 'DR_NDrugs']).agg({'DR_Kol' : 'sum'})
df_xyz
```

```python
df_xyz  = df_xyz.reset_index()
df_xyz = df_xyz[df_xyz['DR_NDrugs'].isin(xyz_names)]. \
    groupby('DR_NDrugs'). \
    apply(lambda x : pd.Series({'Kol' : x['DR_Kol'].std()/x['DR_Kol'].mean()}), include_groups=False). \
    reset_index()
```
