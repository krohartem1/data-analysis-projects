```python
#Мерджим данные
abc = pd.DataFrame({
    'name' : ['Кабачок', 'помидор', 'огурец'],
    'abc' : ['A', 'A', 'C']
})

xyz = pd.DataFrame({
    'name_xyz' : ['Кабачок', 'Баклажан', 'огурец'],
    'xyz' : ['X', 'Y', 'Z']
})
```

```python
abc_xyz = abc.merge(xyz, left_on='name', right_on='name_xyz', how='outer')
abc_xyz
```

```python
#Функция как COALESCE
abc_xyz['name'] = np.where(abc_xyz['name'].isnull(), abc_xyz['name_xyz'], abc_xyz['name'])
abc_xyz
```

```python
#Удаляем ненужный столбец
abc_xyz = abc_xyz.drop('name_xyz', axis=1)
abc_xyz
```

```python
import glob
```

```python
#Нашли все нужные файлы по нашему шаблону
files = glob.glob('D:/GitHub_projects/Simulative_course/extra_excel/*.xlsx')
```

```python
#Читаем много файлов и собираем в один датафрейм\
dfs = []
for file in files:
    small_df = pd.read_excel(file)
    dfs.append(small_df)
```

```python
dfs
```

```python
#Объединяем датафреймы
df = pd.concat(dfs, ignore_index=True)
```

```python
#Полезная функция - iterrows
for i, row in df.iterrows():
    s = f'insert into sales values ("{row['DR_Dat']}", "{row['DR_Kol']}")'
    print(s)
    #print(f'Номер строки: {i}, дата: {row['DR_Dat']}')
    if i == 10:
        break
```
